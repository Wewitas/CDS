##a. Extraer la informacin del archivo

```python
url = 'https://raw.githubusercontent.com/Wewitas/Datasets/refs/heads/main/0.8ddpr74xw650.udno6l076vcountry_vaccinations.csv'
df = pd.read_csv(url)
df.head()
```

#c. Determinar la cantidad de vacunas aplicadas de cada compaa (con base en cmo lo reporta cada pas en la columna vaccines, en otras palabras, agrupe por vaccines y realice la sumatoria).

```python
df[['vaccines', 'daily_vaccinations']].groupby('vaccines').sum()
```

#d. Obtener la cantidad de vacunas aplicadas en todo el mundo.

```python
print('Vacunas aplicadas en el mundo, total de:', f"{int(df['daily_vaccinations'].sum()):,}")
```

#e. Calcular el promedio de vacunas aplicadas por pas.

```python
df[['country', 'daily_vaccinations']].groupby('country').mean()
```

#f. Determinar la cantidad de vacunas aplicadas el da 29/01/21 en todo el mundo.

```python
print('Vacunas aplicadas el 29/01/21:', f"{int(df['daily_vaccinations'][df['date'] == '2021-01-29'].sum()):,}")
```

#g. Crear un dataframe nuevo denominado conDiferencias que contenga los datos originales y una columna derivada (diferencias) con las diferencias de aplicacin entre las columnas daily_vaccionations y daily_vaccionations_raw.  

```python
conDiferencias = df.copy()
conDiferencias['diferencias'] = conDiferencias['daily_vaccinations'] - conDiferencias['daily_vaccinations_raw']
conDiferencias.head()
```

#h. Obtener el periodo de tiempo entre el registro con fecha ms reciente y el registro con fecha ms antigua.

```python
print('Fecha ms reciente', df['date'].max(), '\nFecha ms antigua', df['date'].min())
print('Periodo entre fecha ms antigua y reciente', df['date'].max() - df['date'].min())
```

#i. Crear un dataframe nuevo denominado conCantidad que contenga los datos originales y una columna derivada (canVac) con la cantidad de vacunas utilizadas cada da (usar la columna vaccines y separar por el carcter , ).

```python
conCantidad = df.copy()
conCantidad['canVac'] = conCantidad['vaccines'].str.split(',').str.len()
conCantidad.head()
```

#j. Generar un dataframe denominado antes20 con todos los registros que se hayan realizado antes del 20 de diciembre de 2020.

```python
antes20 = df.copy()
antes20 = antes20[antes20['date'] < '2020-12-20']
print('Cantidad de registro que se realizaron antes del 20/12/20:', antes20.shape[0])  #len(antes20)
antes20.head()
```

#k. Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfizer.

```python
pfizer = df.copy()
pfizer = pfizer[pfizer['vaccines'].str.contains('Pfizer', case=False)] #Se desactiva la sensibilidad a maysculas
print('Cantidad de registro que se realizaron con Pfizer:', pfizer.shape[0])
pfizer.head()
```

#l. Almacenar los dataframes generados (conDiferencias, conCantidad, antes20 y pfizer) en un archivo de Excel denominado resultadosReto.xlsx, donde cada dataframe ocupe una hoja diferente. Se recomienda ver la documentacin de pd.ExcelWriter.

```python
!pip install xlsxwriter
```

```python
with pd.ExcelWriter('resultadosReto.xlsx', engine='xlsxwriter') as writer:
    conDiferencias.to_excel(writer, sheet_name='Diferencias', index=False)
    conCantidad.to_excel(writer, sheet_name='Cantidad', index=False)
    antes20.to_excel(writer, sheet_name='Antes20', index=False)
    pfizer.to_excel(writer, sheet_name='Pfizer', index=False)
```
